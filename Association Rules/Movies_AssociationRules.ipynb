```python
freq_movies = apriori(movies1, min_support = 0.05, use_colnames=True, max_len = 3)
freq_movies
```

```python
rules = association_rules(freq_movies, metric = 'lift', min_threshold= 0.9)
rules
```

```python
# Top 10 rules

rules.sort_values('lift', axis = 0, ascending= False).head(10)
```

```python
rules = association_rules(freq_movies, metric='confidence', min_threshold=0.7)
rules
```

```python
# top 5 rules by confidence

rules.sort_values('confidence', ascending=False).head(5)
```

```python
rules = association_rules(freq_movies, metric='support', min_threshold= 0.2)
rules
```

```python
# top 5 rules by confidence

rules.sort_values('support', ascending=False).head(5)
```

# Visualizing Association Rules

```python
# Support vs Confidence


plt.scatter(rules['support'], rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
# Support vs Lift

plt.scatter(rules['support'], rules['lift'])
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs Confidence')
plt.show()
```

```python
# Lift vs Confidence

plt.scatter(rules['lift'], rules['confidence'])
plt.xlabel('lift')
plt.ylabel('confidence')
plt.title('Lift vs Confidence')
plt.show()
```
